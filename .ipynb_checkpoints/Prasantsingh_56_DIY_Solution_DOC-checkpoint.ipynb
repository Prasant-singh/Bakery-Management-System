```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
import pandas as pd
```

```python
df=pd.read_csv("Groceries data.csv")
df.head()
```

```python
df[df['Member_number']==1001]
```

```python
df["item_count"]=1
```

```python
df
```

```python
df.drop(columns=[ "month","day","year","day_of_week"])
```

```python
df
```

```python
basket = (df.groupby(['Member_number','itemDescription'])['item_count']
          .sum()
          .unstack()
          .reset_index()
          .fillna(0)
          .set_index('Member_number'))
basket
```

```python
def encode_units(x):
    if x <= 0:
        return 0    
    if x >= 1:
        return 1
basket_sets = basket.applymap(encode_units)
basket_sets.head()
```

```python
#Generating frequent item sets that have a support of at least 7%
frequent_itemsets = apriori(basket_sets, min_support=0.07, use_colnames=True)

#Generating the rules with their corresponding support, confidence and lift
rules = association_rules(frequent_itemsets, metric="lift",num_itemsets= int, min_threshold=0.5)

rules.head()
```

```python
#Filtering out the values with lift > = 1 and confidence > = 0.5
rules[ (rules['lift'] >= 1 ) & (rules['confidence'] >= 0.5) ].head()
```
